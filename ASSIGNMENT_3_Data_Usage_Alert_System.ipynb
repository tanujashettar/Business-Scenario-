Business Scenario
Telecom system alerts users based on data usage.

```python
# Input values
total_data = float(input("Enter total data limit (GB): "))
data_used = float(input("Enter data used (GB): "))

# Calculate usage percentage
usage_percent = (data_used / total_data) * 100

# Check usage status
if usage_percent >= 90:
    print("High Usage Alert")
elif usage_percent >= 70 and usage_percent < 90:
    print("Warning")
else:
    print("Normal Usage")
```
