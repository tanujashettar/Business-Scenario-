Business Scenario
The logistics company calculates extra charges based on distance.


```python
# Input distance
distance = int(input("Enter distance in km: "))

# Calculate delivery charge
if distance <= 5:
    charge = 0
elif distance <= 15:
    charge = (distance - 5) * 10
else:
    charge = (10 * 10) + (distance - 15) * 20

# Display total charge
print("Total Delivery Charge: ₹", charge)
```
